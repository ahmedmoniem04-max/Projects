```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```
project_data = pd.read_csv('survey lung cancer.csv')
```

```
project_data.head(10)
```

```
project_data.shape
```

```
project_data.describe()
```

```
print(project_data.isnull().sum())
```

```
plt.figure(figsize=(6, 4))
sns.countplot(x='COUGHING', data=project_data, palette='viridis')
plt.title('COUGHING')
plt.show()
```

```
# Count the occurrences of each of the target values
target_counts = project_data['YELLOW_FINGERS'].value_counts()
# Plot the distribution as the pie chart
plt.figure(figsize=(6, 6))
plt.pie(target_counts, labels=target_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('YELLOW_FINGERS')
plt.show()
```

```
from sklearn.preprocessing import MinMaxScaler
```

```
columns_to_normalize = ['SMOKING', 'ANXIETY', 'AGE']

# Create a Minimum Maximum Scaler
scaler = MinMaxScaler()

# Fit the scaler to the selected columns and transform the data
project_data[columns_to_normalize] = scaler.fit_transform(project_data[columns_to_normalize])

# Display the normalized data
print("\nNormalized Data:")
print(project_data.head())
```

```
correlation_matrix = project_data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectKBest, f_classif

# Assume the X is the feature matrix, and y is the target variable
X = project_data.drop('SMOKING', axis=1)
y = project_data['COUGHING']

# Label encode categorical columns
label_encoder = LabelEncoder()
categorical_columns = X.select_dtypes(include='object').columns
X[categorical_columns] = X[categorical_columns].apply(label_encoder.fit_transform)

# Create the SelectKBest instance
selector = SelectKBest(score_func=f_classif, k=5)

# Fit and transform the feature matrix
X_selected = selector.fit_transform(X, y)

# select feature indices
selected_feature_indices = selector.get_support(indices=True)

# Print the feature names
selected_feature_names = X.columns[selected_feature_indices]
print(f'Selected features: {selected_feature_names}')
```

```
X_selected = X[selected_feature_names]
X_selected
```

```
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)
```

```
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```
logistic_regression = LogisticRegression(random_state=42)

logistic_regression.fit(X_train_scaled, y_train)

y_pred_logistic = logistic_regression.predict(X_test)

accuracy_logistic = accuracy_score(y_test, y_pred_logistic)

classification_report_logistic = classification_report(y_test, y_pred_logistic)
```

```
print("\nLogistic Regression Results:")
print(f"Accuracy: {accuracy_logistic:.4f}")
print("Classification Report:\n", classification_report_logistic)
```

```
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train, y_train)
y_pred_tree = decision_tree.predict(X_test)
accuracy_tree = accuracy_score(y_test, y_pred_tree)
classification_report_tree = classification_report(y_test, y_pred_tree)
```

```
print("\nDecision Tree Results:")
print(f"Accuracy: {accuracy_tree:.4f}")
print("Classification Report:\n", classification_report_tree)
```

```
random_forest = RandomForestClassifier(n_estimators=200)
random_forest.fit(X_train, y_train)
y_pred_forest = random_forest.predict(X_test)
accuracy_forest = accuracy_score(y_test, y_pred_forest)
classification_report_forest = classification_report(y_test, y_pred_forest)
```

```
print("\nRandom Forest Results:")
print(f"Accuracy: {accuracy_forest:.4f}")
print("Classification Report:\n", classification_report_forest)
```

```
non_numerical_columns = project_data.select_dtypes(include='object').columns


df_encoded = pd.get_dummies(project_data, columns=non_numerical_columns)


print(project_data.head())
non_numerical_columns = project_data.select_dtypes(include='object').columns

label_encoder = LabelEncoder()
project_data[non_numerical_columns] = project_data[non_numerical_columns].apply(label_encoder.fit_transform)


print(project_data.head())
```

```
svm = SVC(random_state=42)
svm.fit(X_train_scaled, y_train)
y_pred_svm = svm.predict(X_test_scaled)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
classification_report_svm = classification_report(y_test, y_pred_svm)
```

```
print("\nSVM Results:")
print(f"Accuracy: {accuracy_svm:.4f}")
print("Classification Report:\n", classification_report_svm)
```

```
knn = KNeighborsClassifier( n_neighbors=9)
knn.fit(X_train_scaled, y_train)
y_pred_knn = knn.predict(X_test_scaled)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
classification_report_knn = classification_report(y_test, y_pred_knn)
```

```
print("\nKNN Results:")
print(f"Accuracy: {accuracy_knn:.4f}")
print("Classification Report:\n", classification_report_knn)
```

```
# Define the KNN classifier
knn = KNeighborsClassifier()

# Define the parameter grid for grid search
param_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'p': [1, 2]  # 1 for Manhattan distance, 2 for Euclidean distance
}

# Perform grid search using cross-validation (5-fold cross-validation in this example)
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train_scaled, y_train)

# Get the best parameters and best model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Evaluate the best model on the test set
y_pred = best_model.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)

# Display the results
print("Best Parameters:", best_params)
print("Test Accuracy with Best Model:", accuracy)
```

```
svm = SVC()

param_grid = {
    'C': [0.1, 1, 10, 100],  # Regularization parameter
    'gamma': ['scale', 'auto'],  # Kernel coefficient
    'kernel': ['linear', 'rbf', 'poly']  # Kernel type: linear, radial basis function (RBF), polynomial
}

grid_search = GridSearchCV(svm, param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Perform Grid Search
grid_search.fit(X_train_scaled, y_train)  # Assuming X_train_scaled, y_train are your training data

# Get the best parameters and best model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Evaluate the best model on the test set
y_pred = best_model.predict(X_test_scaled)  # Assuming X_test_scaled is your test data
accuracy = accuracy_score(y_test, y_pred)  # Assuming y_test is your test labels

# Display the results
print("Best Parameters:", best_params)
print("Test Accuracy with Best Model:", accuracy)
```

```
from sklearn.tree import DecisionTreeClassifier

# Define the Decision Tree classifier
decision_tree = DecisionTreeClassifier()

# Define the parameter grid for grid search
param_grid_dt = {
    'max_depth': [3, 5, 7, 9],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Perform grid search using cross-validation (5-fold cross-validation in this example)
grid_search_dt = GridSearchCV(decision_tree, param_grid_dt, cv=5, scoring='accuracy')
grid_search_dt.fit(X_train_scaled, y_train)

# Get the best parameters and best model
best_params_dt = grid_search_dt.best_params_
best_model_dt = grid_search_dt.best_estimator_

# Evaluate the best model on the test set
y_pred_dt = best_model_dt.predict(X_test_scaled)
accuracy_dt = accuracy_score(y_test, y_pred_dt)

# Display the results for Decision Tree
print("Best Parameters for Decision Tree:", best_params_dt)
print("Test Accuracy with Best Decision Tree Model:", accuracy_dt)
```
